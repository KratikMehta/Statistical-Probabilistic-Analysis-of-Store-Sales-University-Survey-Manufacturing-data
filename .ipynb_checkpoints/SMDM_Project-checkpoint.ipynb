Coefficient of Variations is a measure of relative dispersion around the mean. It is used to compare variables with different ranges and units of measurments.

```python
region_vs_items_cv = region_vs_items_std / region_vs_items_mean
region_vs_items_cv
```

    * In general, spendings for all the items are largely spread out around the mean.

# Problem 2

The Student News Service at Clear Mountain State University (CMSU) has decided to gather data about the undergraduate students that attend CMSU. CMSU creates and distributes a survey of 14 questions and receives responses from 62 undergraduates (stored in the __Survey__ data set).

```python
survey_df = pd.read_csv('Survey-1.csv')
survey_df.head()
```

## Checking the types of variables in the dataset.

```python
survey_df.info()
```

From the above output we can see that:

## 2.1. For this data, construct the following contingency tables (Keep Gender as row variable)

```python
gender_vs_employment = pd.crosstab(index=survey_df['Gender'], columns=survey_df['Employment'], margins=True, margins_name='Total')
gender_vs_employment
```

## 2.3. Assume that the sample is representative of the population of CMSU. Based on the data, answer the following question:

```python
print(f"Conditional probability that a Male student has an Accounting major is {round(account_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has a CIS major is {round(cis_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has an Economics/Finance major is {round(economics_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has an International Business major is {round(business_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has a Management major is {round(manage_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has Other major is {round(other_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has a Retailing/Marketing major is {round(market_male_prob, 2)}%.")
print(f"Conditional probability that a Male student has an Undecided major is {round(undecided_male_prob, 2)}%.")
```

### 2.3.2. Find the conditional probability of different majors among the female students in CMSU.

```python
account_female_prob = gender_vs_major.loc['Female', 'Accounting'] / gender_vs_major.loc['Female', 'Total'] * 100
cis_female_prob = gender_vs_major.loc['Female', 'CIS'] / gender_vs_major.loc['Female', 'Total'] * 100
economics_female_prob = gender_vs_major.loc['Female', 'Economics/Finance'] / gender_vs_major.loc['Female', 'Total'] * 100
business_female_prob = gender_vs_major.loc['Female', 'International Business'] / gender_vs_major.loc['Female', 'Total'] * 100
manage_female_prob = gender_vs_major.loc['Female', 'Management'] / gender_vs_major.loc['Female', 'Total'] * 100
other_female_prob = gender_vs_major.loc['Female', 'Other'] / gender_vs_major.loc['Female', 'Total'] * 100
market_female_prob = gender_vs_major.loc['Female', 'Retailing/Marketing'] / gender_vs_major.loc['Female', 'Total'] * 100
undecided_female_prob = gender_vs_major.loc['Female', 'Undecided'] / gender_vs_major.loc['Female', 'Total'] * 100
```

## 2.4. Assume that the sample is a representative of the population of CMSU. Based on the data, answer the following question:

### 2.4.1. Find the probability that a randomly chosen student is a male and intends to graduate.

```python
grad_yes_male = gender_vs_grad_intent.loc['Male', 'Yes'] / 62 * 100
print(f"Out of the 62 students, 17 students are male who intends to graduate.\n"
      f"Hence the probability that a randomly chosen student is a male and intends to graduate is {round(grad_yes_male, 2)}%.")
```

## 2.5. Assume that the sample is representative of the population of CMSU. Based on the data, answer the following question:

### 2.5.1. Find the probability that a randomly chosen student is a male or has full-time employment?

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $$

```python
full_time_male = (gender_vs_employment.loc['Male', 'Total'] + 
                  gender_vs_employment.loc['Total', 'Full-Time'] - 
                  gender_vs_employment.loc['Male', 'Full-Time']) / 62 * 100
print(f"Out of the 62 students, 29 are male, 10 have full time jobs, and 7 are both male and have full time jobs.\n"
      f"Hence the probability that a randomly chosen student is a male or has full-time employment is {round(full_time_male, 2)}%.")
```

### 2.5.2. Find the conditional probability that given a female student is randomly chosen, she is majoring in international business or management.

For mutualy exclusive events: $$ P(A \cup B \mid C) = P(A \mid C) + P(B \mid C) $$

```python
business_manage_female = gender_vs_major.loc['Female', ['International Business', 'Management']].sum() / gender_vs_major.loc['Female', 'Total'] * 100
print(f"Out of the 33 female students, 4 are majoring in International Business and 4 are majoring in Management.\n"
      f"Hence the conditional probability that given a female student is randomly chosen, she is majoring in "
      f"international business or management is {round(business_manage_female, 2)}%.")
```

## 2.6.  Construct a contingency table of Gender and Intent to Graduate at 2 levels (Yes/No). The Undecided students are not considered now and the table is a 2x2 table. Do you think the graduate intention and being female are independent events?

```python
gender_vs_intent_2_levels = survey_df.loc[survey_df['Grad Intention'] != 'Undecided', ['Gender', 'Grad Intention']]
gender_vs_intent_2_levels = pd.crosstab(index=gender_vs_intent_2_levels['Gender'], 
                                        columns=gender_vs_intent_2_levels['Grad Intention'], 
                                        margins=True, margins_name='Total')
gender_vs_intent_2_levels
```

For independent events: $$ P(A \cap B) = P(A)P(B) $$

```python
female_prob = gender_vs_intent_2_levels.loc['Female', 'Total'] / gender_vs_intent_2_levels.loc['Total', 'Total']
yes_intent_prob = gender_vs_intent_2_levels.loc['Total', 'Yes'] / gender_vs_intent_2_levels.loc['Total', 'Total']
yes_intent_and_female = gender_vs_intent_2_levels.loc['Female', 'Yes'] / gender_vs_intent_2_levels.loc['Total', 'Total']
print("Out of the 40 students, 20 are male, 28 intent to graduate, and 11 are both female and intent to graduate.\n")

if yes_intent_and_female == (female_prob * yes_intent_prob):
    print("Hence using the above mentioned formula we find that the graduate intention and being female are independent events.")
else:
    print("Hence using the above mentioned formula we find that the graduate intention and being female are not independent events.")
```

## 2.7. Note that there are four numerical (continuous) variables in the data set, GPA, Salary, Spending, and Text Messages.<br/>Answer the following questions based on the data

## 2.8. Note that there are four numerical (continuous) variables in the data set, GPA, Salary, Spending, and Text Messages. For each of them comment whether they follow a normal distribution. Write a note summarizing your conclusions., marker='o', fmt='b'

```python
survey_df.hist(column=['GPA', 'Salary', 'Spending', 'Text Messages'], 
                  figsize=(12, 6), bins=8, layout=(2, 2))
plt.figtext(s='Distribution of Continous Variables in Survey dataset', x=0.5, y=1, ha='center', size='xx-large')
plt.tight_layout();
```

Theoretically the mean, median and mode of a normal distribution are equal.

### Distribution of GPA

```python
gpa_mean = survey_df['GPA'].mean()
gpa_median = survey_df['GPA'].median()
gpa_mode = survey_df['GPA'].mode()
gpa_std = survey_df['GPA'].std()
print(f"For the GPA variable:\n"
      f"Mean = {round(gpa_mean, 1)}\n"
      f"Median = {round(gpa_median, 1)}\n"
      f"Mode = [{gpa_mode[0]}, {gpa_mode[1]}, {gpa_mode[2]}]")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
sm.qqplot(survey_df['GPA'], line='45', loc=gpa_mean, scale=gpa_std, ax=ax, marker='o', fmt='b')
plt.title('QQ-Plot for GPA');
```

From the above outputs we can conclude that the __GPA__ variable in the Survey dataset approximately __follows a normal distribution__. This can also be verified from the above histogram.

### Distribution of Salary

```python
salary_mean = survey_df['Salary'].mean()
salary_median = survey_df['Salary'].median()
salary_mode = survey_df['Salary'].mode()
salary_std = survey_df['Salary'].std()
print(f"For the Salary variable:\n"
      f"Mean = {round(salary_mean, 1)}\n"
      f"Median = {round(salary_median, 1)}\n"
      f"Mode = {salary_mode[0]}")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
sm.qqplot(survey_df['Salary'], line='45', loc=salary_mean, scale=salary_std, ax=ax, marker='o', fmt='b')
plt.title('QQ-Plot for Salary');
```

The mean, median and mode of Salary are very different from each other but its QQ-Plot is close to the straight line. Hence the __Salary__ variable approximately __follows a normal distribution__. This can also be verified from the above histogram.

```python
spend_mean = survey_df['Spending'].mean()
spend_median = survey_df['Spending'].median()
spend_mode = survey_df['Spending'].mode()
spend_std = survey_df['Spending'].std()
print(f"For the Spending variable:\n"
      f"Mean = {round(spend_mean, 1)}\n"
      f"Median = {round(spend_median, 1)}\n"
      f"Mode = {spend_mode[0]}")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
sm.qqplot(survey_df['Spending'], line='45', loc=spend_mean, scale=spend_std, ax=ax, marker='o', fmt='b')
plt.title('QQ-Plot for Spending');
```

Even though the mean, median and mode of Spending are close to each other, its QQ-Plot varies a lot from the straight line.Hence the __Spending__ variable __does not follows a normal distribution__. This can also be verified from the above histogram.
